###**Observation**
<ul><li>Country has only one value , <b>United States</b>; we will be drpping the country column.</li></ul> 

```python
data = data.drop(columns=['Country','Postal Code'] , axis=1)
data.head()
```

##**Data Visualization**

```python
#history distribution of numeric features
data.hist(bins=20,figsize=(18,16),)
```

##**Heatmap**

```python
#correlation between features
corr = data.corr()
corr
```

```python
# Visualizing correlation between features
plt.figure(figsize=(15,9))
sns.heatmap(corr,annot=True,center=1)
plt.show()
```

##**Observation**
<ul><li>From the heatmap we can see here a positive cor-relation between sales and profit that is <b>0.48</b> , thus <b>Profit increases ,simultanously sales increases.</b>
<li>And here we can see a negative cor-relation between profit an discount is <b>-0.22 </b> and sales and discount of <b>-0.028</b>, thus<b> discount increases , simultanously profit goes down .

```python
#calculating statewise profit
profit_state_g = data.groupby('State')['Profit'].sum()
profit_state_g
```

##**State-wise profit of US**

```python
#State-wise profit of US
plt.subplots(figsize=(15,17),)
profit_state_g.plot(kind='barh', fontsize=15,color='darkslategrey')
plt.xlabel('profit',fontsize=15,)
plt.ylabel('state',fontsize=13)
plt.title('statewise profit',fontsize=14)
plt.show()
```

###**Observation**
<ul><li>Highest profit is generated by <b>California</b> and <b>Newyork</b>.</li>
<li>Lowest profit is generated by <b>Texas</b> and <b>ohlo</b></li>
</ul>
<ul><ul>By the subplot we can clearly see that:
     <ul><li>Top 5 states giving profit are <b>California , Newyork , Washington and michigan .</b></li></ul>
     <ul><li>Top 5 states giving loss are <b>Texas , Ohio , lllinois , pennsylvania.</b></li></ul>

```python
#calculating statewise sales
Sales_state_g = data.groupby('State')['Sales'].sum()
Sales_state_g
```

```python
#State wise Sales of US
plt.figure(figsize=(18,12))
sns.barplot(x=Sales_state_g.index, y=Sales_state_g.values , palette='rocket')
plt.ylabel("Sales")
plt.xticks(rotation=80)
plt.show()
```

###**Observation**
<ul><li>Thus giving max profit California also has most sales.</li></ul>
<ul><ul>we can see that the Lowest sales are from North Dakota<li>By the subplot we can signify that</li></ul></ul>
<ul><li>Top 5 states giving most of the sales are from <b> California , Newyork, Washington and Texas.</b></li></ul>
<ul><li>Top 5 states giving loss are <b>North Dakota , Maine , South Dakota, West Virginica.</b></li></ul>

```python
#Comparing Sales and Profit and Statewise
sales_Profit = data.groupby('State')['Sales','Profit'].sum()
sales_Profit.plot.bar(figsize=(14,9),)
plt.show()
```

###**Observation**
<ul><li>State <b>California, Washington, Newyork, michigan</b> have <b>highest sales and profit.</b></li>
<li>State <b> Texas, Pennsylvania, Ohio, lllinois</b> have <b> higher sales but profit is negative (in loss).</b></li></ul>

```python
data.Discount.head(9)
```

```python
#State wise average Discount of US
avg_dis = np.round(data.groupby('State')['Discount'].mean()*100,decimals=0)
avg_dis
```

```python
#State wise avg_discount of US
plt.figure(figsize=(15,9))
sns.barplot(x=avg_dis.index,y=avg_dis.values,palette='Set1')
plt.xticks(rotation=80)
plt.show()
```

###**Observation**
<ul><li>lllinois and Texas has highest average discount among other states of US.</li>
<li>We can also see that here Top 5 states having most discount are lllinois, Texas, Pennsylvania, Ohlo and Colorado.</li></ul>

```python
# Calculating category wise sales, Discount, Profit
cat_data = data.groupby(['Category'])[['Sales','Discount','Profit']].mean()
cat_data
```

```python
#plotting donut chart to see category wise sales, Discount, Profit
ex=[0.05,0.05,0.05]
cat_data.plot.pie(subplots=True,figsize=(19,20),autopct='%0.1f%%',label=cat_data.index,pctdistance=0.65,explode=ex)
circle = plt.Circle(xy=(0,0),radius=0.50,facecolor='white')
plt.text(0,0,'category',horizontalalignment='center',verticalalignment='center')
plt.gca().add_artist(circle)
plt.show()
```

##**Observation**
<ul><li><b>Technology</b> category has the <b>maximum Sales and profit and less discount compare to furniture and office.</b></li>
<li><b>Office supplies and Furniture</b> category has <b>minimum sales and Profit</b> as compare to Technology category.</li>

```python
# Calculating Ship Mode wise sales, Discount, Profit
ship_data = data.groupby(['Ship Mode'])[['Sales','Discount','Profit']].mean()
ship_data
```

```python
#plotting donut chart to see category wise sales, Discount, Profit
ex=[0.05,0.05,0.05,0.05]
ship_data.plot.pie(subplots=True,figsize=(19,20),autopct='%0.1f%%',label=cat_data.index,pctdistance=0.65,explode=ex)
#circle
circle = plt.Circle(xy=(0,0),radius=0.50,facecolor='white')
plt.text(0,0,'Ship Mode',horizontalalignment='center',verticalalignment='center')
plt.gca().add_artist(circle)
plt.show()
```

###**Observation**
<ul><li>In Ship Mode, <b>Same Day</b> has the<b> highest sales.</b></li>
<li><b>First class</b> has the <b>highest Profit and Discount.</b></li>
<li>In Ship Mode,<b> Satandard Class</b> has <b>lowest sales as well as lowest profit.</b></li>

```python
# Comparing Sales Sub-Category Wise
sub_cat = data.groupby('Sub-Category')['Sales'].mean()
# Sub_cat
plt.figure(figsize=(15,9))
sns.barplot(x=sub_cat.index,y=sub_cat.values,palette='viridis')
plt.xticks(rotation=80)
plt.show()
```

###**Observation**
<ul><li>Sub-Category items <b>Copier, Machines and Table</b> have the <b>highest Sales.</b>

```python
# Comparing Sales Sub-Category Wise
sub_cat = data.groupby('Sub-Category')['Profit'].sum()
# Sub_cat
plt.figure(figsize=(15,9))
sns.barplot(x=sub_cat.index,y=sub_cat.values,palette='icefire_r')
plt.xticks(rotation=80)
plt.show()
```

###**Observation***
<ul><li>We can see here Sub-Category , items<b> Copiers has the highest Sales as well as highest profit margin.</li></b>
<li>We can see here <b>Machine had second highest sales but less profit margin</b>, due to giving <b>large amount of discount.</b></li>
<li>In sub-category, item <b>Table had third highest sales</b> but<b> it's profit is in negative (on loss).</b></li>
